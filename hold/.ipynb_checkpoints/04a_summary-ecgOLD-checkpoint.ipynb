# Visualize the data 
## 1. Import packages and load data
Load the .csv file that contains the interval-related analysis results from neurokit2. 

```python
import os
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Paths to folders
results_folder = '/Users/seinj/Teaching/Results/'

# Load the .csv file containing the summary of the results
filename = results_folder + 'ecg_results.csv'
results = pd.read_csv(filename)

print(results)
```

## 2. Select parameters and conditions of interest
First we will look at the mean ECG rate parameter, the number of heartbeats per minute.   
We will aggregate over participants rather than contrasting between them, as commonly done in full experiments. This means that we are not interested in how sub-01 differs from sub-02 in a given condition, but rather in how walk_ground and walk_high conditions differ, for example.    
That is why there is one vector (or list) per condition, which would include more data points than 2 if we had more data sets. 

```python
# Filter rows for each condition
base_ground = results[results['Condition'] == 'base_ground']
base_high = results[results['Condition'] == 'base_high']
walk_ground = results[results['Condition'] == 'walk_ground']
walk_high = results[results['Condition'] == 'walk_high']

# Select the column containing the parameter of interest
bpm_base_ground = base_ground['ECG_Rate_Mean']
bpm_base_high = base_high['ECG_Rate_Mean']
bpm_walk_ground = walk_ground['ECG_Rate_Mean']
bpm_walk_high = walk_high['ECG_Rate_Mean']

# Show data 
print(bpm_base_ground)
print(bpm_base_high)
print(bpm_walk_ground)
print(bpm_walk_high)
```

You will notice that the base high condition is missing for one participant, resulting in imbalanced data. Normally you would exclude that participant from analysis, but for demonstration purpose we will use a fake number to fill the missing data in.  

```python
pd.concat([bpm_base_high, pd.Series([75])], ignore_index=True)
```

## 3. Use divisive baseline to compute base to walk ratio

```python
# Calculate the ratio of walk to baseline for both conditions
# Note that this only works because in the series, the order of participants match (sub-01 comes before sub-02)
ratio_ground = bpm_walk_ground.values / bpm_base_ground.values
ratio_high = bpm_walk_high.values / bpm_base_high.values

# Calculate the mean ratio for each condition
avg_ratio_ground = ratio_ground.mean()
avg_ratio_high = ratio_high.mean()
```

## 4. Visualize
### 4.1. Visualize all data using basic bar plot
We now visualize the average base to walk ratios computed per condition. First, let us have a look at what it looks like when we use a basic bar plot from matplotlib.   

Note how the **unit** and the **axis labels** are defined. These are essential information to be included in a plot. 

```python
# Data for the bar plot
conditions = ['Ground (Walk/Base)', 'High (Walk/Base)']
averages = [avg_ratio_ground, avg_ratio_high]

# Create the bar plot
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(conditions, averages, color=['skyblue', 'orange'])

# Add labels and title
ax.set_ylabel('Average Ratio (Walk/Base)')
ax.set_title('Average Heart Rate Ratio: Walk to Base')

# Display the plot
plt.show()
```

### 4.2. Better visualization
Pick a better color, display the scale more clearly by making some changes. There are loads of online resources for good visualization tutorials. Choosing the right color is also important for communication of your results.
```python
data_points_ground = [ratio for ratio in ratio_ground] 
data_points_high = [ratio for ratio in ratio_high] 

# Create the bar plot
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(conditions, averages, color=['skyblue', 'orange'], width=0.3)

# Add scatter plot for individual data points
ax.scatter([0] * len(data_points_ground), data_points_ground, color='skyblue', label='Ground Data Points', zorder=5)
ax.scatter([1] * len(data_points_high), data_points_high, color='orange', label='High Data Points', zorder=5)

# Add scatter plot for individual data points with jitter and matching colors
#ax.scatter(np.zeros(len(data_points_ground)), data_points_ground, color='#0073e6', label='Ground', zorder=5)
#ax.scatter(np.ones(len(data_points_high)), data_points_high, color='#f57600', label='High', zorder=5)

# Add labels and title
ax.set_ylabel('Average Ratio (Walk/Base)')
ax.set_title('Average Heart Rate Ratio: Walk to Base')

# Add a legend in the upper right corner
ax.legend(loc='upper right', bbox_to_anchor=(1.8, 1))

# Add grid lines
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, color='gray', alpha=0.7)

# Display the plot and save
figure_filename = results_folder + 'ecg_summary.png'
plt.tight_layout()
plt.subplots_adjust(right=0.55)
plt.savefig(figure_filename, bbox_inches='tight', pad_inches=0.1)
plt.show()
```
