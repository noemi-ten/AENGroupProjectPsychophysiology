```python
import os
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

results_folder = '/content/drive/MyDrive/Project Psychophysiology/spider/results/'

# Load the .csv file containing the summary of the results
filename = results_folder + 'eda_results.csv'
results = pd.read_csv(filename)

print(results.head())

# Filter rows for each condition
baseline = results[results['Condition'] == 'baseline']
spiderhand = results[results['Condition'] == 'spiderhand']
spidervideo = results[results['Condition'] == 'spidervideo']

# Select the column containing the parameter of interest
scr_baseline = baseline['SCR_Peaks_N']
scr_spiderhand = spiderhand['SCR_Peaks_N']
scr_spidervideo = spidervideo['SCR_Peaks_N']

# Show data
print(scr_baseline)
print(scr_spiderhand)
print(scr_spidervideo)

# Calculate the ratio of different tasks to baseline
ratio_spiderhand = scr_spiderhand.values / scr_baseline.values
ratio_spidervideo = scr_spidervideo.values / scr_baseline.values

# Calculate the mean ratio for each task
avg_ratio_spiderhand = ratio_spiderhand.mean()
avg_ratio_spidervideo = ratio_spidervideo.mean()

# Data for the bar plot
tasks = ['Spiderhand', 'Spidervideo']
averages = [avg_ratio_spiderhand, avg_ratio_spidervideo]

# Create the bar plot
fig, ax = plt.subplots(figsize=(6, 4))
bars = ax.bar(tasks, averages, color=['skyblue', 'orange'], width=0.3)

# Extract individual data points
data_points_spiderhand = [ratio for ratio in ratio_spiderhand]
data_points_spidervideo = [ratio for ratio in ratio_spidervideo]

# Add scatter plot for individual data points with jitter and matching colors
ax.scatter(np.zeros(len(data_points_spiderhand)), data_points_spiderhand, color='#0073e6', label='Spiderhand', zorder=5)
ax.scatter(np.ones(len(data_points_spidervideo)), data_points_spidervideo, color='#f57600', label='Spidervideo', zorder=5)

# Add labels and title
ax.set_ylabel('Average SCR Ratio')
ax.set_title('Ratio SCR: Tasks to Baseline')

# Add a legend in the upper right corner
ax.legend(loc='upper right', bbox_to_anchor=(1.8, 1))

# Add grid lines
ax.grid(True, which='both', axis='y', linestyle='--', linewidth=0.7, color='gray', alpha=0.7)

# Display the plot and save
figure_filename = results_folder + 'eda_summary.png'
plt.tight_layout()
plt.subplots_adjust(right=0.55)
plt.savefig(figure_filename, bbox_inches='tight', pad_inches=0.1)
plt.show()

print(avg_ratio_spiderhand)
print(avg_ratio_spidervideo)



```
